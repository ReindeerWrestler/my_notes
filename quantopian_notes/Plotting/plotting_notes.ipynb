```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pandas_datareader as pdr

data_nvda = pdr.get_data_yahoo('NVDA')

data_t = pdr.get_data_yahoo('T')
data_t.head()
```

```python
plt.hist(data_t['Open'], bins=20)
plt.xlabel('Open Price')
plt.ylabel('Number of Days Observed')
plt.title('Frequency Distribution of AT&T Prices');
```

# Returns Histogram

```python
# Remove the first element because percent change from nothing to something is NaN
R = data_t['Open'].pct_change()[1:]

# Plot a histogram using 20 bins
plt.hist(R, bins=20)
plt.xlabel('Return')
plt.ylabel('Number of Days Observed')
plt.title('Frequency Distribution of AT&T Returns');
```

# Cumulative Histogram (Discrete Estimated CDF)

```python
plt.hist(R, bins=20, cumulative=True)
plt.xlabel('Return')
plt.ylabel('Number of Days Observed')
plt.title('Cumulative Distribution of AT&T Returns');
```

# Scatter plot

```python
plt.scatter(data_t['Open'], data_nvda['Open'])
plt.xlabel('AT&T')
plt.ylabel('NVDA')
plt.title('Daily Prices');
```

```python
R_nvda = data_nvda['Open'].pct_change()[1:]
R_t = data_t['Open'].pct_change()[1:]

plt.scatter(R_nvda, R_t)
plt.xlabel('NVDA')
plt.ylabel('AT&T')
plt.title('Daily Returns');
```
