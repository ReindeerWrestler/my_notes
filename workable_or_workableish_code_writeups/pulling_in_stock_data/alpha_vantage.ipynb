docs for alpha vantage api usage

```python
import json
import pandas as pd
from alpha_vantage.timeseries import TimeSeries

f = open("api_keys.json")
api_keys = json.load(f)

ts = TimeSeries(key=api_keys["alpha_vantage"])

data, meta_data = ts.get_daily(symbol="NVDA")
print(data)

```

data in pandas format

```python
ts = TimeSeries(key=api_keys["alpha_vantage"], output_format='pandas')

data, meta_data = ts.get_daily(symbol="NVDA")
print(data)
```

using pandas datareader

```python
import pandas_datareader as pdr
# Get Alpha Vantage Data
data = pdr.get_data_alphavantage("NVDA", api_key=api_keys["alpha_vantage"])
# Summarize
print(data.info())
```
